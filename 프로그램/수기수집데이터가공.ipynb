```python
import pandas as pd
import os
import traceback
```

```python
# 인코딩 이상, 헤더 한 칸(정상)
def utf_changer(path):
    df = pd.read_csv(path, encoding = 'cp949')
    df.to_csv(path, encoding = 'utf-8-sig', index = False)
    return df
# 인코딩 이상, 헤더 두 칸
def make_header_2to1(path):
    df = pd.read_csv(path, encoding = 'cp949')
    list_columns = list(df.columns)
    for index, i in enumerate(list_columns):
        if 'Unnamed' not in i:
            replacing_column = i
        else :
            list_columns[index] = replacing_column
    final_list = []
    for index, i in enumerate(list(df.loc[0])):
        final_list.append(str(list_columns[index]) + '_' + str(i))
    for index, i in enumerate(final_list):
        i = i.replace('_nan', '')
        final_list[index] = i
    df.loc[0] = final_list
    df.columns = df.iloc[0]
    df.drop(index = 0, axis = 0, inplace = True)
    df.reset_index(drop = True, inplace = True)
    df.to_csv(path, encoding = 'utf-8-sig', index = False)
    return df
# 인코딩 이상, 헤더 세 칸
def make_header_3to1(path):
    df = pd.read_csv(path, encoding = 'cp949')
    list_columns = list(df.columns)
    for index, i in enumerate(list_columns):
        if 'Unnamed' not in i:
            replacing_column = i
        else :
            list_columns[index] = replacing_column
    list_columns2 = list(df.loc[0])
    for index, i in enumerate(list_columns2):
        list_columns2[index] = str(i)
    for index, i in enumerate(list_columns2):    
        if i != 'nan':
            replacing_column = i
        elif i == 'nan' and index != 0 :
            try:
                list_columns2[index] = replacing_column
            except:
                pass
    final_list = []
    for index, i in enumerate(list(df.loc[1])):
        final_list.append(str(list_columns[index]) + '_' + str(list_columns2[index]) + '_' + str(i))
    for index, i in enumerate(final_list):
        i = i.replace('_nan', '')
        final_list[index] = i
    df.loc[0] = final_list
    df.columns = df.iloc[0]
    df.drop(index = [0,1], axis = 0, inplace = True)
    df.reset_index(drop = True, inplace = True)
    df.to_csv(path, encoding = 'utf-8-sig', index = False)
    return df

def read_filelist(path):
    try:
        print('#### Read path {}'.format(path))
        file_list = list([])
        
        for dirpath, _, filenames in os.walk(path):
            for f in filenames:
                try:
                    file_list.append(os.path.abspath(os.path.join(dirpath, f)))
                except:
                    print('######## Read file error : {}'.format(filenames))
                    print('############ {}'.format(traceback.format_exc()))
                
        return file_list
    except:
          
        print('#### Read file list error')
        print('######## {}'.format(traceback.format_exc()))
```
