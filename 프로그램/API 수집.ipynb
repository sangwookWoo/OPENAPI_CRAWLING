```python
import requests
import pprint
import json
import pandas as pd
from pandas.io.json import json_normalize
```

# 계약종별 전력 사용량

```python
df = pd.read_csv('시도코드.csv')
years = range(2002,2023)
months = range(10,13)
df_va = pd.DataFrame(columns = ['year','month','metro','city','cntr','custCnt','powerUsage','bill','unitCost'])
api_key = 'api_key'
for year in years:
    for month in months:
        for code in df['코드']:
            try :
                url = f'https://bigdata.kepco.co.kr/openapi/v1/powerUsage/contractType.do?year={year}&month={month}&metroCd={code}&apiKey={api_key}&returnType=json'
                response = requests.get(url)
                contents = response.text
                json_ob = json.loads(contents)
                data = json_ob['data']
                dataframe = pd.json_normalize(data)
                df_va = pd.concat([df_va, dataframe], axis = 0)
            except :
                print(json_ob, year, month)
df_va.to_csv('계약종별전력사용량.csv', encoding = 'utf-8-sig', index = False)
```
